```python
import sys
sys.path.append("/anaconda/lib/python2.7/site-packages")

import numpy as np
import pandas as pd
import os
```

```python
rent = pd.read_csv("data/neighborhood-lines/housing_prices_by_neighborhood.csv")
```

```python
info = pd.read_csv("data/neighborhood-lines/neighborhood_info.csv")
```

```python
boroughs = ["Brooklyn", "Bronx", "Manhattan", "Queens", "Staten Island"]
```

```python
rent.Date.values
```

```python
rent.as_matrix()
```

```python
df = pd.DataFrame(rent.as_matrix()[:, 1:].T, columns=rent.Date.values)
df["neighborhood"] = rent.columns[1:]
```

```python
info.head()
```

MEDIAN

```python
median_df = pd.DataFrame(rent["Date"])

for borough in boroughs:
    test = info[info['borough'] == borough]

    sub_df = test.merge(df, on="neighborhood").drop(info.columns, axis=1)

    medians = []
    for col in sub_df.columns:
        vals = sub_df[col].values
        median = np.median(vals)
        medians.append(median)
    
    median_df[borough] = medians
    
median_df.to_csv("data/neighborhood-lines/median_rent_new.csv", index=False)
```

```python
rent.head()
```

MEAN

```python
mean_df = pd.DataFrame(rent["Date"])

for borough in boroughs:
    test = info[info['borough'] == borough]

    sub_df = test.merge(df, on="neighborhood").drop(info.columns, axis=1)

    means = []
    for col in sub_df.columns:
        vals = sub_df[col].values
        mean = np.mean(vals)
        means.append(mean)
    
    mean_df[borough] = means

mean_df.to_csv("data/neighborhood-lines/mean_rent_new.csv", index=False)
```
